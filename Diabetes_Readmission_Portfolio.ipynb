### ğŸ“Š **Executive Summary**
This portfolio presents a comprehensive analysis of diabetes patient readmission patterns using machine learning techniques. We analyzed **101,766 patient encounters** from 130 US hospitals (1999-2008) to predict 30-day readmission risk and identify key factors driving readmissions.

### ğŸ¯ **Key Findings**
- **Overall Readmission Rate**: 11.16% (11,357 out of 101,766 patients)
- **Model Performance**: XGBoost achieved 67.02% ROC-AUC, Logistic Regression 66.22% ROC-AUC
- **Top Risk Factors**: Number of medications, time in hospital, and previous emergency visits
- **Demographic Insights**: 20-30 age group shows highest readmission risk (14.24%)

```python
# Import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set style for beautiful plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Configure plotly
import plotly.io as pio
pio.templates.default = "plotly_white"

print("ğŸ“Š All libraries imported successfully!")
print("ğŸ¨ Ready to create beautiful visualizations!")
```

```python
# Load and prepare the data
print("ğŸ”„ Loading diabetes dataset...")
df = pd.read_csv('diabetic_data.csv')

# Create target variable
df['target'] = (df['readmitted'] == '<30').astype(int)

# Clean data
df = df.dropna(subset=['race', 'gender', 'age'])

print(f"âœ… Dataset loaded successfully!")
print(f"ğŸ“ˆ Dataset shape: {df.shape}")
print(f"ğŸ¯ Target distribution: {df['target'].value_counts()}")
print(f"ğŸ“Š Readmission rate: {df['target'].mean():.2%}")
```

## ğŸ“ˆ **Dataset Overview**

Our analysis is based on a comprehensive dataset containing patient demographics, medical history, and hospital stay information. The dataset provides a rich foundation for understanding readmission patterns.


```python
# Create a beautiful overview dashboard
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Dataset Size', 'Readmission Distribution', 'Age Distribution', 'Gender Distribution'),
    specs=[[{"type": "indicator"}, {"type": "pie"}],
           [{"type": "bar"}, {"type": "bar"}]]
)

# Dataset size indicator
fig.add_trace(go.Indicator(
    mode = "number",
    value = len(df),
    title = {"text": "Total Patients"},
    number = {"font": {"size": 40, "color": "#2E86AB"}}
), row=1, col=1)

# Readmission distribution pie chart
readmit_counts = df['target'].value_counts()
fig.add_trace(go.Pie(
    labels=['No Readmission', 'Readmitted <30 days'],
    values=readmit_counts.values,
    hole=0.4,
    marker_colors=['#FF6B6B', '#4ECDC4']
), row=1, col=2)

# Age distribution
age_counts = df['age'].value_counts().sort_index()
fig.add_trace(go.Bar(
    x=age_counts.index,
    y=age_counts.values,
    marker_color='#45B7D1',
    name='Age Groups'
), row=2, col=1)

# Gender distribution
gender_counts = df['gender'].value_counts()
fig.add_trace(go.Bar(
    x=gender_counts.index,
    y=gender_counts.values,
    marker_color='#96CEB4',
    name='Gender'
), row=2, col=2)

fig.update_layout(
    title_text="ğŸ¥ Diabetes Readmission Dataset Overview",
    title_x=0.5,
    title_font_size=24,
    showlegend=False,
    height=600
)

fig.show()
```
